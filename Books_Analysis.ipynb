```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Books_Data_Clean.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df=df[df["Publishing Year"]>1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset="Book Name", inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df["Publishing Year"], color='c')
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
df["genre"].value_counts().plot(kind="bar", color='m')
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in Each Genre")
plt.show()
                                
                                
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x="genre", y="Book_ratings_count", data=df, hue="genre", palette='rainbow', legend=False)
plt.xlabel("Genre")
plt.ylabel("Book_ratings_count")
plt.title("Box Plot of Book Rating for Each Genre")
plt.show()
```

```python
plt.scatter(df['sale price'],df['units sold'], c='r')
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter plot of Sale price VS Units Sold")
plt.show()
```

```python
language_count=df["language_code"].value_counts()
```

```python
plt.pie(language_count, labels=language_count.index, startangle=90, autopct="%1.1f%%")
plt.title("Language Distribution of Books")
plt.show()
```

```python
df.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending=False).max()
```

```python
df.groupby("language_code").size().sort_values(ascending=False)
```

```python
plt.scatter(df["Book_average_rating"],df["Book_ratings_count"], color='c')
plt.xlabel("Book Average Rating")
plt.ylabel("Book Ratings Count")
plt.title("Scatter Plot Book Avg Rating VS Book Rating Count")
plt.show()
```

```python
total_gross_sales_by_author=df.groupby("Author")["gross sales"].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending=False).head(20).plot(kind="bar", color='y')
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sale for each Author")
plt.show()
```

```python
sns.boxplot(x="Author_Rating", y="units sold", data=df, hue="Author_Rating", palette="rainbow")
plt.xlabel("Author Rating")
plt.ylabel("Unit SOld")
plt.title("Box Plot of Unit Sold by Each Author Rating")
plt.show()
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line", marker="o")
plt.xlabel("Publishing Year")
plt.ylabel("Total Units Sold")
plt.title("Total Units Sold Over the Years")
plt.show()
```
